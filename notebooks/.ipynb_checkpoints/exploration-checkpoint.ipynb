```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All packages imported successfully!")
```

```python
# Load your generated dataset
df = pd.read_csv("../data/simulated_candidates.csv")

# Check the distribution of performance labels
print(df["performance_label"].value_counts())
```

```python
# Quick look at numeric features
print(df.describe())

# Check unique education levels
print(df["education_level"].value_counts())
```

```python
#Visual representation of simulated target
sns.countplot(x="performance_label", data=df)
plt.title("Distribution of Candidate Performance Labels")
plt.show()
```

```python
# Check distribution of numeric features
df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
# Showing features influencing performance score
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Boxplot for all scores
# Features to visualize
features = ["technical_test_score", "interview_score", "personality_score", "years_experience", "previous_rating"]

for feature in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x="performance_label", y=feature, data=df)
    plt.title(f"{feature} vs Performance Label")
    plt.show()
```

## Summary of Observations
- Years of experience has the __strongest correlation__ with performance_score
- Technical, interview, and personality scores contribute but are less dominant
- Education level shows minor influence
- High performers are rare, mimicking real recruitment distributions
- __No missing values__, all features properly scaled
